```python
new_stock_df = pd.read_csv('./Datos/dimension_stock_item_new.csv', sep=',')
new_stock_df.head(20)
```

```python
for col in new_stock_df.columns:
    nan = new_stock_df[col].isna().sum()
    print("La columna %s tiene %i valores nulos" % (col, nan))
```

```python
new_stock_df.drop(["Brand"], axis=1, inplace=True)
new_stock_df["Color"].fillna("Unknown", inplace=True)
new_stock_df["Size_val"].fillna("Unknown", inplace=True)
new_stock_df.head(20)
```

```python
for col in new_stock_df.columns:
    nan = new_stock_df[col].isna().sum()
    print("La columna %s tiene %i valores nulos" % (col, nan))
```

```python
new_stock_df['Tax_Rate'] = [x.replace(',','.') for x in new_stock_df['Tax_Rate']]
new_stock_df['Unit_Price'] = [x.replace(',','.') for x in new_stock_df['Unit_Price']]
new_stock_df['Recommended_Retail_Price'] = [x.replace(',','.') for x in new_stock_df['Recommended_Retail_Price']]
new_stock_df['Typical_Weight_Per_Unit'] = [x.replace(',','.') for x in new_stock_df['Typical_Weight_Per_Unit']]
```

```python
new_stock_df.head(20)
```

```python
new_stock_df = new_stock_df.dropna()
new_stock_df.head(20)
```

```python
stock_historia_df = stock_df
```

```python
stock_historia_df['tk_stock_item'] = stock_historia_df.index
stock_historia_df['_version'] = 1
stock_historia_df['date_from'] = datetime.datetime(1900, 1, 1)
stock_historia_df['date_to'] = datetime.datetime(2199, 12, 31)
```

```python
stock_historia_df
```

```python
for row in new_stock_df.iterrows():
    stock_item_key = row[1]["Stock_Item_Key"]
    
    version = 0
    # ahora se recorre el antiguo dataframe
    indice = 0
    size = stock_historia_df.size
    indice_real = 0
    usar_indice_real = False
    for row2 in stock_historia_df.iterrows():
        key = row2[1]["stock_item_key"]
        if stock_item_key == key:
            ver = row2[1]["_version"] # version
            if(ver > version):
                version = ver
                indice_real = indice
            usar_indice_real = True
        indice+=1
    
    # hay que detectar si hubo algún cambio respecto a la version final
    if( usar_indice_real and
        (row[1]["Stock_Item"] != stock_historia_df.loc[indice_real]["stock_item"] or
        row[1]["Color"] != stock_historia_df.loc[indice_real]["color"] or
        row[1]["Selling_Package"] != stock_historia_df.loc[indice_real]["selling_package"] or
        row[1]["Buying_Package"] != stock_historia_df.loc[indice_real]["buying_package"] or
        row[1]["Size_val"] != stock_historia_df.loc[indice_real]["size_val"] or
        row[1]["Lead_Time_Days"] != stock_historia_df.loc[indice_real]["lead_time_days"] or
        row[1]["Quantity_Per_Outer"] != stock_historia_df.loc[indice_real]["quantity_per_outer"] or
        row[1]["Is_Chiller_Stock"] != stock_historia_df.loc[indice_real]["is_chiller_stock"] or
        row[1]["Tax_Rate"] != stock_historia_df.loc[indice_real]["tax_rate"] or
        row[1]["Unit_Price"] != stock_historia_df.loc[indice_real]["unit_price"] or
        row[1]["Recommended_Retail_Price"] != stock_historia_df.loc[indice_real]["recommended_retail_price"] or
        row[1]["Typical_Weight_Per_Unit"] != stock_historia_df.loc[indice_real]["typical_weight_per_unit"] ) ):
        
            dt = datetime.datetime.now()
            stock_historia_df.loc[indice_real,['date_to']] = datetime.datetime(dt.year, dt.month, dt.day)
            row_a_agregar = stock_historia_df.loc[indice_real]
            row_a_agregar["_version"] = version + 1
            dt = datetime.datetime.now()
            
            row_a_agregar["stock_item"] = row[1]["Stock_Item"]
            row_a_agregar["color"] = row[1]["Color"]
            row_a_agregar["selling_package"] = row[1]["Selling_Package"]
            row_a_agregar["buying_package"] = row[1]["Buying_Package"]
            row_a_agregar["size_val"] = row[1]["Size_val"]
            row_a_agregar["lead_time_days"] = row[1]["Lead_Time_Days"]
            row_a_agregar["quantity_per_outer"] = row[1]["Quantity_Per_Outer"]
            row_a_agregar["is_chiller_stock"] = row[1]["Is_Chiller_Stock"]
            row_a_agregar["tax_rate"] = row[1]["Tax_Rate"]
            row_a_agregar["unit_price"] = row[1]["Unit_Price"]
            row_a_agregar["recommended_retail_Price"] = row[1]["Recommended_Retail_Price"]
            row_a_agregar["typical_Weight_Per_Unit"] = row[1]["Typical_Weight_Per_Unit"]
            row_a_agregar["tk_stock_item"] = size + 1
            size += 1
            
            row_a_agregar["date_from"] = datetime.datetime(dt.year, dt.month, dt.day)
            row_a_agregar["date_to"] = datetime.datetime(2199, 12, 31)
            stock_historia_df = stock_historia_df.append(row_a_agregar, ignore_index=True)
            #stock_historia_df = pd.concat([stock_historia_df, row_a_agregar], axis = 1)
```

```python
stock_historia_df.sort_values(by=['stock_item_key'])
stock_historia_df.head(20)
```

```python
stock_historia_df.to_sql('stockitem_historia',con=conexion,if_exists='append',index=False)
```
